```python
# splitting dataset into training set and test set
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.20, random_state = 0)
```

```python
# creating K-NN classifier model and training the model
from sklearn.neighbors import KNeighborsClassifier
classifier = KNeighborsClassifier(n_neighbors = 5, metric = 'minkowski', p = 2)
classifier.fit(x_train, y_train)
```

```python
# predicting species
y_pred = classifier.predict(x_test)
np.set_printoptions(precision = 2)
print("Predicted values are:")
print(y_pred)
```

```python
numrows = len(y_test)
```

```python
# creating new dataframe to store Original_species and Predicted_species
regression_df = pd.DataFrame({'Original_species':[0.0]*numrows,
                              'Predicted_species':[0.0]*numrows})
```

```python
seq = ['Original_species',
       'Predicted_species']
```

```python
regression_df = regression_df.reindex(columns=seq)
```

```python
for i in range(0 , numrows):
    real_values = y_test[i]
    predicted_values = y_pred[i]
    
    if real_values == "setosa":
        real_values = 0
    elif real_values == "versicolor":
        real_values = 1
    else:
        real_values = 2
        
    
    if predicted_values == "setosa":
        predicted_values = 0
    elif predicted_values == "versicolor":
        predicted_values = 1
    else:
        predicted_values = 2    
    
    regression_df.iloc[i,0] = real_values
    regression_df.iloc[i,1] = predicted_values


regression_df = regression_df.dropna() #exclude any rows with missing data
print(regression_df)
```

```python
# replacing 0 with "setosa" and storing the updated dataframe into dataframe df1
df1 = regression_df.replace(to_replace =0, value ="setosa")
```

```python
df1
```

```python
# replacing 1 with "versicolor" and storing the updated dataframe into dataframe df2
df2 = df1.replace(to_replace =1, value ="versicolor")
```

```python
df2
```

```python
# replacing 2 with "virginica" and storing the updated dataframe into dataframe df3
df3 = df2.replace(to_replace =2, value ="virginica")
```
