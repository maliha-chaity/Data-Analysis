```python
# importing libraries
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
# importing datasets
dataset_train = pd.read_csv('train.csv')
dataset_test = pd.read_csv('test.csv')
```

```python
# Dropping unnecessary columns from training datatest
dataset_train.drop(dataset_train.columns[[0,1,5,7,9,10,11,12]], axis = 1, inplace = True)
```

```python
# Dropping columns with no values from training dataset
df_train = dataset_train.dropna()
```

```python
# Replace with Yes with 1, No with 0 in Married column of training dataset
df_train = df_train.replace({'Married': {'Yes': 1, 'No': 0}})
```

```python
# Replace Graduate with 1, Not Graduate with 0 in Education column in training dataset
df_train = df_train.replace({'Education': {'Graduate': 1, 'Not Graduate': 0}})
```

```python
# Replace 3+ with 4 in Dependents column of training dataset for making removing calculation problems
df_train = df_train.replace({'Dependents': {'3+': 4}})
```

```python
# Dropping unnecessary column from test dataset
dataset_test.drop(dataset_test.columns[[0,1,5,7,9,10,11]], axis = 1, inplace = True)
```

```python
dataset_test
```

```python
# Dropping rows with no values from test dataset
df_test = dataset_test.dropna()
```

```python
df_test
```

```python
# Replace Yes with 1, No with 0 in Married column of test dataset
df_test = df_test.replace({'Married': {'Yes': 1, 'No': 0}})
```

```python
df_test
```

```python
# Replace Graduate with 1, Not Graduate 0 in Education dataset of test dataset
df_test = df_test.replace({'Education': {'Graduate': 1, 'Not Graduate': 0}})
```

```python
# Replace 3+ with 4 in Dependents column of test dataset for removing problem in calculations
df_test = df_test.replace({'Dependents': {'3+': 4}})
```

```python
df_test
```

```python
x_train = df_train.iloc[:, :-1].values
y_train = df_train.iloc[:, -1].values
```

```python
x_test = df_test.iloc[:, :-1].values
y_test = df_test.iloc[:, -1].values
```

```python
# Training the model
from sklearn.linear_model import LinearRegression
regressor = LinearRegression()
regressor.fit(x_train, y_train)
```

```python
# making prediction after training the model
y_pred = regressor.predict(x_test)
np.set_printoptions(precision = 2)
```

```python
numrows = len(y_test)
```

```python
# Creating new dataframe which only contains the actual loan amount and the predicted loan amount
regression_df = pd.DataFrame({'Original_Loan_Amount':[0.0]*numrows,
                              'Predicted_Loan_Amount':[0.0]*numrows})
```

```python
seq = ['Original_Loan_Amount',
       'Predicted_Loan_Amount']
```

```python
regression_df = regression_df.reindex(columns=seq)
```

```python
for i in range(0 , numrows):
    real_values = y_test[i]
    predicted_values = y_pred[i]
    
    regression_df.iat[i,0] = real_values
    regression_df.iat[i,1] = predicted_values


regression_df = regression_df.dropna() #exclude any rows with missing data
print(regression_df)
```

```python
# print the actual and predicted loan amount in a new csv file
regression_df.to_csv('predict_loan.csv', index = True) 
```
